# AMS-02 Proton Flux Data from PRL 2021: Daily Spectra and Event Selection

In this project, we are working with data related to cosmic rays, specifically focusing on the daily proton flux data collected by the Alpha Magnetic Spectrometer (AMS-02) experiment. 

The dataset presented here comprises the daily proton fluxes from the Alpha Magnetic Spectrometer (AMS-02) as published in the Physical Review Letters (PRL) 2021, focusing on the proton component. This unique dataset offers an opportunity to explore the temporal evolution of cosmic radiation at 1 astronomical unit (AU), representing Earth's orbit around the Sun in interstellar space.

These spectra were measured aboard the International Space Station, situated within Earth's magnetosphere, but meticulously filtered to exclude any influence from the geomagnetic field. This meticulous event selection process is detailed in the article by M. Aguilar et al. (AMS Collaboration) in Phys. Rev. Lett. 114, 171103 (2015). Out of approximately 1.5 x 10^11 cosmic ray events recorded from May 20, 2011, to October 29, 2019, this dataset includes 5.5 x 10^9 protons.

To assess the magnetosphere's impact on cosmic rays, a helpful resource is the COR system web tool, accessible at https://cor.crmodels.org/.

Data Description:
The dataset includes cosmic ray proton spectra collected by the AMS-02 experiment [1], featuring proton fluxes in a CSV file:

**p_AMS_PRL2021_daily.csv**: This data table contains AMS-02 daily proton fluxes collected from May 20, 2011, to October 29, 2019, spanning the rigidity interval from 1 to 100 GV. The data is sourced from the Cosmic ata.(C## B) [2].**

```python
import numpy as np
import pandas as pd
```

```python
# read the csv data
df = pd.read_csv('p_AMS_PRL2021_daily.csv')
df.head()
```

See the range of rigidities specified in the second and third columns. The energetic bins employed by AMS-02 are represented in GigaVolts (GV), distributed across 30 rigidity bins, encompassing values from 1.0 GV to 100 GV.

Rigidity, denoted as R for protons, is defined as:

R = pc / q

Here, 'p' signifies particle momentum, 'c' denotes the speed of light, and 'q' represents the elementary charge. The kinetic energy, T_kin, usually expressed in Giga electronVolts (GeV) units, can be computed from rigidity using the following equation:

T_kin = âˆš(T_0^2 + R^2 - T_0)

In this equation, 'T_0' corresponds to the proton's rest energy (0.938272 GeV), and rigidity 'R' is measured in GV.

### Spectra visualization

```python
import matplotlib.pyplot as plt

plt.scatter(df['rigidity_min GV'],df['proton_flux m^-2sr^-1s^-1GV^-1'],c=df['proton_flux m^-2sr^-1s^-1GV^-1']) #c=df['proton_flux m^-2sr^-1s^-1GV^-1']: This part sets the color of the data points in the scatter plot based on the values in the 'proton_flux m^-2sr^-1s^-1GV^-1' column. 

plt.xscale("log")
plt.yscale("log")

plt.xlabel('Rigidity [GV]')
plt.ylabel('proton flux $m^{-2}sr^{-1}s^{-1}GV^{-1}$')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.rcParams['font.size'] = 14
```

```python
# now creating 3 new columns for year, months and days

df['year'] = df['date YYYY-MM-DD'].dt.year
df['month'] = df['date YYYY-MM-DD'].dt.month
df['day'] = df['date YYYY-MM-DD'].dt.day

# Check the DataFrame
df.head()
```

Visualization of all daily spectra in one 3D figure.


```python
from mpl_toolkits import mplot3d #This submodule provides tools for creating 3D plots and visualizations (3D scatter plots, surface plots, etc.)
import matplotlib.ticker

fig = plt.figure(figsize= (15,14), dpi=300, edgecolor = 'red') #dots per inch (dpi)
# syntax for 3-D projection
ax = plt.axes(projection = '3d')

# time, approximation for 3D visualisation, first version, in later upades will be used more precise version
df['timeTemp'] = df.year + ((df.month-1)/12) + ((df.day - 1)/365.25 )
df['logrig'] = np.log10(df['rigidity_min GV'])
df['logint'] = np.log10(df['proton_flux m^-2sr^-1s^-1GV^-1'])

ax.scatter(df.timeTemp, df.logrig, df.logint, c = df.timeTemp, s = 2)

# setting the view or perspective of a 3D plot, azimuthal angle (azim) of the view (determines the angle at which you view the 3D plot)
ax.view_init(azim = 30)
ax.set_xlabel('Time', fontsize = 10)
ax.set_ylabel('Rigidity GV', fontsize = 10)
ax.set_zlabel('Proton Flux $m^{-2}sr^{-1}s^{-1}GV^{-1}$', fontsize = 10)

majors = [-2.0,-1.0,0.0,1.0,2.0,3.0,4.0]
ax.zaxis.set_major_locator(matplotlib.ticker.FixedLocator(majors))
labels = [0.01,0.1,1,10,100,1000,10000]
ax.set_zticklabels(labels)

majors = [0.0,1.0,2.0]
ax.yaxis.set_major_locator(matplotlib.ticker.FixedLocator(majors))
labels = [1,10,100]
ax.set_yticklabels(labels)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
font = {'size': 10}
ax.tick_params('z', labelsize=font['size'])
plt.rcParams['font.size'] = 10
```
